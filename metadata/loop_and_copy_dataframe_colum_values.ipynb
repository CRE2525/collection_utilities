```python
import numpy as np
import pandas as pd
```

```python
def lookup_fn(df, key_col, key_val, val_col, idx=0):
    try:
        result_df = df.loc[df[key_col] == key_val].iloc[0]
        value = result_df[val_col]
        return value
    except IndexError:
        return 0
```

```python
df1 = pd.read_csv('data/docs2people_flatten.csv') 
df1.head(10)
```

```python
df1["related_people_original"] = ""
```

```python
df2 = pd.read_csv('data/docs2people_needs_review.csv') 
df2.head(2)
```

```python
for index, row in df1.iterrows():
    document_id = str(row["Correct Document ID"])
    action = str(row["Update_Person_Note_1"])
    if action == "parse":
        related_people_original = lookup_fn(df2, "document_id", document_id, "related_people", idx=0)
        df1.set_value(index, 'related_people_original', related_people_original)
```

```python
df1.to_csv("report/rel_people.csv", index=False)
```
